## 2. Variables
---

Kenapa kita butuh variable?


- Variabel adalah lokasi memori yang dicadangkan untuk menyimpan nilai-nilai.
- Kaidah Penulisan variabel Python:

    - Karakter pertama harus berupa huruf atau garis bawah/underscore _
    - Tidak boleh ada spasi
    - Karakter selanjutnya dapat berupa huruf, garis bawah/underscore _ atau angka
    - Karakter pada nama variabel bersifat sensitif (case-sensitif).
    - Tidak bisa menggunakan reserved words (and, exec, not, finally, for, or, while break etc.). [Lihat Selengkapnya](https://www.w3schools.com/python/python_ref_keywords.asp)
    
    
    
- Metode Penulisan variabel Python:
    - Snake Case : `nama_variable`
    - Camel Case : `namaVariable`
    - Pascal Case : `NamaVariable`

### 4.1 List

List dalam bahasa Python adalah daftar nilai yang dipisahkan koma (item) antara tanda kurung siku. Hal penting tentang daftar adalah item dalam list tidak harus sama jenisnya.

Anggota sebuah list diidentifikasi menggunakan index, index sendiri dimulai dari 0.

Contoh List:
```python
list1 = ['PAJAK', 'AKUTANSI', 1993, 2017]
list2 = [1, 2, 3, 4, 5 ]
list3 = ["a", "b", "c", "d"]
```


Terdapat beberapa method yang umum berkaitan dengan list

- ```len(list)```     : Mengetahui panjang list 

- ```append(item)```  : Menambahkan sebuah item ke dalam list

- ```extend(list)```  : Mengisi anggota list dengan anggota dari list lain

- ```remove(item)```  : Menghapus sebuah item ke dalam list

- ```count(item)```   : Menghitung keberadaan sebuah item pada sebuah list

- ```index(item)```   : Mengetahui posisi sebuah item pada sebuah list

- ```sort()```  : Mengurutkan item-item dalam list

- ```clear()```   : Mengosongkan nilai list

- ```pop(index)```    : Menghapus anggota list pada posisi tertentu

- ```reverse()``` : Membalikan urutan anggota list
    

#### 4.1.1 Mengakses anggota List

Untuk mengakses anggota list, gunakan indeks. Indeks dimulai dari 0 (artinya untuk mendapatkan nilai pertama sebuah list kita menggunakan indeks 0. 

Contoh:
```python
mata_kuliah = ['Akuntansi','Perpajakan','Makro Ekonomi','Statistik']
mata_kuliah[0] # 'Akuntansi'
```
Menggunakan index negatif berarti mengambil urutan dari belakang
```python
mata_kuliah[-1] # 'Statistik'

```

#### 4.1.2 List Slicing

_Slicing_ adalah membagi atau mengambil bagian dari list berdasarkan _range_ indeks tertentu. 

Cara _slicing_ ditulis sebagai berikut:

```python 
    daftar = [1,2,3,4]
    daftar[start:end:step]
```
Note: end bersifat eksklusif. start,end,step tidak harus didefinisikan 

Contoh:
```python 
    daftar = [1,2,3,4,5,6,7,8,9]
    daftar[1:6]
    daftar[4:]
    daftar[1:6:2]
    daftar[-5:-1]
```

## 5. Fungsi
Fungsi adalah blok kode terorganisir dan dapat digunakan kembali yang digunakan untuk melakukan sebuah tindakan/action. 




Mendefinisikan fungsi dengan Python.

- Fungsi blok dimulai dengan  kata kunci `def` diikuti oleh nama fungsi dan tanda kurung (()).
- Tata cara penamaan fungsi sama seperti dengan variabel.
- Argumen/Parameter dalam sebuah fungsi adalah input dari luar fungsi untuk dapat diproses oleh fungsi
- `Return` mengembalikan nilai dari proses yang dilakukan


````python
def square(x):
  return x**2
````

````python
square(2)
````

### 4.1 Fungsi Lambda

Kita dapat membuat suatu fungsi sederhana yang mengembalikan nilai dalam satu baris menggunakan operator lambda. 

````python
# Contoh Lambda Function
fx = lambda x: x**2 + 5*x + 6
fx(4)

# Contoh Lambda Function
luas_segitiga = lambda alas, tinggi: alas * tinggi / 2
luas_segitiga(4, 7)
````
